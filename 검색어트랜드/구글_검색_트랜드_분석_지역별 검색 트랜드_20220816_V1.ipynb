```python
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import os
```

```python
# 검색 Keyword, 검색기간 입력
keyword = "WTO"
period = "now 7-d" #최근 7일 
```

```python
 # Goolge Trend 접속

trend_obj = TrendReq()
trend_obj.build_payload(kw_list=[keyword], timeframe=period)
```

```python
# 지역별 검색 trend 비교 1
trend_df = trend_obj.interest_by_region()
print(trend_df.head())
```

```python
# 지역별 검색 trend 비교 2
# Order by Interest , Desc 
trend_df =trend_df.sort_values("WTO", ascending=False)
print(trend_df.head())
```

```python
trend_df.iloc[:50, :]
```

```python
# 그래프 출력
plt.style.use('ggplot')
plt.figure(figsize=[14,10])
trend_df.iloc[:50, :].plot(kind='bar')
plt.title("Google Trends by Region", size=15)
plt.xticks(fontsize= 8)
plt.legend(labels=[keyword], loc= "upper right")
plt.savefig("Trends by Region", dpi=300)
```
