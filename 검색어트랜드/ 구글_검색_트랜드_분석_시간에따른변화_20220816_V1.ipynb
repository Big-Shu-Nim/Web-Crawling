```python
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import os
```

```python
#검색 키워드, 검색기간 입력

keyword = "Galaxy Fold"
period =  "today 3-m" # 검색기간 : 최근 3개월 
```

```python
# 구글 트랜드 접속
trend_obj = TrendReq()
trend_obj.build_payload(kw_list=[keyword], timeframe = period) 
# 사용된 메쏘드 역할  검색단어와 검색기간 전달 
```

```python
# 시간에 따른 Trend 변화
trend_df = trend_obj.interest_over_time() # 사용된 메쏘드 역할: 시간에 따른 트랜드 분석 로드
print(trend_df.head())
```

```python
# 그래프 출력
plt.style.use("ggplot")
plt.figure(figsize=(14,5))
trend_df["Galaxy Fold"].plot()
plt.title("Goolge Trends over Time", size=15)
plt.legend(labels=["Galaxy Fold"], loc="upper right")
```

```python
output=plt.style.use("ggplot")
plt.figure(figsize=(14,5))
trend_df["Galaxy Fold"].plot()
plt.title("Goolge Trends over Time", size=15)
plt.legend(labels=["Galaxy Fold"], loc="upper right")
plt.savefig("Trends over time", dpi=300)
```
